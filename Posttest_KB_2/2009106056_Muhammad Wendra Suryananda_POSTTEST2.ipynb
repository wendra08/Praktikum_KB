## Menampilkan preview data
- Jumlah record
- nama attribute
- Jumlah attribute dengan tipe objek
- Jumlah attribute dengan tipe angka
- Jumlah attribute dengan nilau null


membuat variabel jumlah record untuk dictionary

```python
jumlahr = len(data)
```

membuat variabel nama attribute untuk dictionary

```python
namaatt = data.columns
```

membuat variabel untuk membuat tipe objek didictionary

```python
datao = len(data.select_dtypes(include=['object']).columns)
```

membuat variabel untuk membuat tipe data angka didictionary

```python
dataiandf = len(data.select_dtypes(include=['int64','float64']).columns)
```

membuat variabel untuk data null

```python
datanull = len(data.columns[data.isna().sum() > 0])
```

membuat dictionary

```python
dictio = {
    "jumlahrecord": jumlahr,
    "namaattribute": namaatt.to_list(),
    "dataobject": datao,
    "dataintdanfloat": dataiandf,
    "datanull": datanull
}
```

Menampilkan dictionary

```python
pd.Series(dictio)
```
