#### Mengelompokkan berdasarkan jumlah dari attribute status

```python
data_pie = dataset_Life_Expectancy['Status'].value_counts().rename_axis('Status').reset_index(name='Id_count')
```

### pltfigure berguna untuk mengubah ukuran output gambar
- colors mengubah warna 
- explode berguna untuk memisahkan salah satu bagan di pie chart
- plt pie maksudnya ngambil dari variabel data_pie yang sudah dideklarasikan di atas dengan labels yg diambil adalah status
- autopct untuk menampilkan persentase, shadow memberikan bayangan, startangle untuk merotasi grafik pie sebanyak 90 derajat
- plt tittle memberikan judul


```python
plt.figure(figsize=(5,5))
colors = sns.color_palette('bright')
explode = [0.1,0]
plt.pie(data_pie['Id_count'], labels=data_pie.Status,
autopct='%1.1f%%', shadow=True, startangle=90, explode=explode)
plt.title("Data Persentase status negara")
plt.show()
```

- dari data persentase status negara, negara berkembang lebih banyak jumlahnya dari negara maju, karena hampir seluruh negara di dunia adalah negara berkembang, contohnya Indonesia
